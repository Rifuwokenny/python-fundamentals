# Lists and Dictionaries

```python
marks = [65, 72, 80, 58, 90]
print(marks)
```

```python
print(marks[0])
print(marks[2])
print(marks[-1])
```

```python
for mark in marks:
    print(mark)
```

```python
print("Highest", max(marks))
print("Lowest", min(marks))
print("Average", sum(marks) / len(marks))
```

```python
marks.append(75)
print(marks)
```

```python
marks.remove(58)
print(marks)
```

```python
marks = [63, 34, 41, 75, 96]
for mark in marks:
    if mark > 50:
        print(mark)
```

```python
students = [
    { "name": "Junior", "marks": 60 },
    { "name": "Kenny", "marks": 45},
    { "name": "Rifuwo", "marks": 80} ]
for student in students:
    if student["marks"] >= 50:
        print(student["name"], "has passed")
    else:
        print(student["name"], "has failed")
```

```python
students = [ 
    { "name": "Junior", "marks": 60 },
    { "name": "Kenny", "marks": 45 },
    { "name": "Rifuwo", "marks": 70 } ]

passed_count = 0
failed_count = 0

for student in students:
    if student["marks"] >= 50:
        print(student["name"], "has passed")
        passed_count += 1
    else:
        print(student["name"], "has failed")
        failed_count += 1

        print("\nSummary:")
        print("Total Students:", len(students))
        print("Passed:", passed_count)
        print("Failed:", failed_count)
```
